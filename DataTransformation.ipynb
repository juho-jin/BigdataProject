## 2. Data Extraction

### 2-1. DateFrame Merging

```python
car_num = '5221'
log_dir = './data/logdata/'
log_files = os.listdir(log_dir+car_num)
log_files
```

```python
pro_dir = './data/prodata/'
pro_files = os.listdir(pro_dir+car_num)
pro_files
```

```python
logd = []
prod = []
for i in range(0,len(pro_files)):
    ldr = log_dir+car_num+'/'+log_files[i]
    pdr = pro_dir+car_num+'/'+pro_files[i]
    logd.append(pd.read_csv(ldr, encoding='cp949'))
    prod.append(pd.read_csv(pdr, encoding='cp949'))
```

```python
logd[0].tail()
```

```python
prod[0].tail()
```

```python
for i in range(0,len(logd)):
    logd[i]['RPMS'] = prod[i]['RPMS']
    logd[i]['ACC'] = prod[i]['ACC']
    logd[i]['FS'] = prod[i]['FS']
    logd[i]['RPMSR'] = prod[i]['RPMSR']
    logd[i]['ACCR'] = prod[i]['ACCR']
    logd[i]['SRA'] = prod[i]['SRA']
```

```python
nsdf = []
for n in range(0,len(logd),1):
    tdf = pd.DataFrame(columns=['date', 'model_name', 'car_type', 'car_num', 'business_num',
       'driver_code', 'daily_distance', 'cum_distance', 'information_date',
       'car_speed', 'RPM', 'brake_signal', 'car_location_GPS_X',
       'car_location_GPS_Y', 'Global_Positioning_System_GPS_azimuth',
       'accelerationVx', 'accelerationVy', 'status_code', 'RPMS', 'ACC', 'FS',
       'RPMSR', 'ACCR', 'SRA'])
    for m in range(0,len(logd[n]),1):
        inda = logd[n]['information_date'][m]
        inda = [int(a) for a in str(inda)]
        if (inda[11] == 0):
            od = logd[n].loc[m]
            tdf.loc[len(tdf)] = [od[0],od[1],od[2],od[3],od[4],
                                od[5],od[6],od[7],od[8],od[9],
                                od[10],od[11],od[12],od[13],od[14],
                                od[15],od[16],od[17],od[18],od[19],
                                od[20],od[21],od[22],od[23]]
        fs = logd[n]['FS'][m]
        rp = logd[n]['RPMSR'][m]
        ac = logd[n]['ACCR'][m]
        sa = logd[n]['SRA'][m] 
        if (fs != 0 or rp != 0 or ac != 0 or sa != 0):
            od = logd[n].loc[m]
            if inda[11] == 1:
                continue
            tdf.loc[len(tdf)] = [od[0],od[1],od[2],od[3],od[4],
                                od[5],od[6],od[7],od[8],od[9],
                                od[10],od[11],od[12],od[13],od[14],
                                od[15],od[16],od[17],od[18],od[19],
                                od[20],od[21],od[22],od[23]]
    nsdf.append(tdf)
```

### 2-2. DataFrame Editing

```python
car_num = '9582'
full_dir = './data/fulldata/'
full_files = os.listdir(full_dir+car_num)
full_files
```

```python
fdfs = []
for cs in full_files[:]:
    fdfs.append(pd.read_csv(full_dir+car_num+'/'+cs, encoding='cp949'))
fdfs[0].tail()
```

```python
fdfs[0].columns
```

```python
fdfs[0][['car_location_GPS_X','car_location_GPS_Y']]
```

```python
for g in range(0,len(fdfs),1):
    for a in range(0,len(fdfs[g]),1):
        ndx = fdfs[g]['car_location_GPS_X'][a]
        ndx = str(ndx)
        ndx = ndx[:3]+'.'+ndx[3:]
        fdfs[g]['car_location_GPS_X'][a] = ndx
        ndy = fdfs[g]['car_location_GPS_Y'][a]
        ndy = str(ndy)
        ndy = ndy[:2]+'.'+ndy[2:]
        fdfs[g]['car_location_GPS_Y'][a] = ndy
    
```
